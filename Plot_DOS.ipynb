## **Non spin-polarised DOS**

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to read the DOS data from a .dat file
def read_dos_data(filename):
    # Assuming the .dat file has two columns: energy and DOS
    data = np.loadtxt(filename)
    energy = data[:, 0]
    dos = data[:, 1]
    return energy, dos

# Function to plot the DOS data with customizable parameters
def plot_dos(
    energy, dos,
    font_size=12, line_thickness=2, frame_thickness=1.5,
    figure_size=(8, 6), x_limits=None, y_limits=None
):
    # Set figure size
    plt.figure(figsize=figure_size)

    # Plot the data
    plt.plot(energy, dos, linewidth=line_thickness)

    # Customize the plot
    plt.xlabel('$E-E_F$ (eV)', fontsize=font_size)
    plt.ylabel('Density of States', fontsize=font_size)
    plt.title('Density of States', fontsize=font_size)

    # Set axis limits if provided
    if x_limits:
        plt.xlim(x_limits)
    if y_limits:
        plt.ylim(y_limits)

    # Customize tick parameters
    plt.tick_params(axis='both', which='major', labelsize=font_size)

    # Customize frame thickness
    plt.gca().spines['top'].set_linewidth(frame_thickness)
    plt.gca().spines['bottom'].set_linewidth(frame_thickness)
    plt.gca().spines['left'].set_linewidth(frame_thickness)
    plt.gca().spines['right'].set_linewidth(frame_thickness)

    # Show the plot
    plt.show()

# Example usage
filename = 'TDOS.dat'  # Replace with your .dat file name
energy, dos = read_dos_data(filename)

# Customize the plot parameters as needed
plot_dos(
    energy, dos,
    font_size=14, line_thickness=2.5, frame_thickness=2,
    figure_size=(10, 7), x_limits=(-3, 3), y_limits=(0, 3)
)
```

## **Spin polarised DOS**

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to read spin-polarized DOS data from a .dat file
def read_spin_polarized_dos_data(filename):
    # Assuming the .dat file has three columns: energy, DOS_up, DOS_down
    data = np.loadtxt(filename)
    energy = data[:, 0]
    dos_up = data[:, 1]
    dos_down = data[:, 2]
    return energy, dos_up, dos_down

# Function to plot the spin-polarized DOS data with customizable parameters
def plot_spin_polarized_dos(
    energy, dos_up, dos_down,
    font_size=12, line_thickness=2, frame_thickness=1.5,
    figure_size=(8, 6), x_limits=None, y_limits=None
):
    # Set figure size
    plt.figure(figsize=figure_size)

    # Plot the spin-up and spin-down DOS
    plt.plot(energy, dos_up, label='Spin Up', color='blue', linewidth=line_thickness)
    plt.plot(energy, dos_down, label='Spin Down', color='red', linewidth=line_thickness)

    # Customize the plot
    plt.xlabel('$E-E_F$ (eV)', fontsize=font_size)
    plt.ylabel('Density of States', fontsize=font_size)
    plt.title('Spin-Polarized Density of States', fontsize=font_size)
    plt.axhline(0, color='black', linewidth=0.5)  # To show the zero line for DOS

    # Set axis limits if provided
    if x_limits:
        plt.xlim(x_limits)
    if y_limits:
        plt.ylim(y_limits)

    # Customize tick parameters
    plt.tick_params(axis='both', which='major', labelsize=font_size)

    # Customize frame thickness
    plt.gca().spines['top'].set_linewidth(frame_thickness)
    plt.gca().spines['bottom'].set_linewidth(frame_thickness)
    plt.gca().spines['left'].set_linewidth(frame_thickness)
    plt.gca().spines['right'].set_linewidth(frame_thickness)

    # Add legend
    plt.legend(fontsize=font_size)

    # Show the plot
    plt.show()

# Example usage
filename = 'TDOS.dat'  # Replace with your .dat file name
energy, dos_up, dos_down = read_spin_polarized_dos_data(filename)

# Customize the plot parameters as needed
plot_spin_polarized_dos(
    energy, dos_up, dos_down,
    font_size=14, line_thickness=2.5, frame_thickness=2,
    figure_size=(10, 7), x_limits=(-8, 8), y_limits=(-5, 5)
)
```

## **Partial Density of states - single element**

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to read partial DOS data from a .dat file
def read_partial_dos_data(filename):
    # Assuming the .dat file has the following columns:
    # energy, s, p_x, p_y, p_z, d_xy, d_yz, d_z2, d_xz, d_x2_y2
    data = np.loadtxt(filename)
    energy = data[:, 0]
    orbitals = data[:, 1:]  # Extract all orbital columns
    return energy, orbitals

# Function to plot selected orbitals from the partial DOS data for both spin channels
def plot_spin_polarized_partial_dos(
    energy_up, orbitals_up, energy_down, orbitals_down,
    selected_orbitals, labels, colors, font_size=12,
    line_thickness=2, frame_thickness=1.5, figure_size=(10, 6),
    x_limits=None, y_limits=None
):
    # Set figure size
    plt.figure(figsize=figure_size)

    # Plot the selected orbitals for spin-up and spin-down
    for index, orbital_index in enumerate(selected_orbitals):
        # Spin-up
        plt.plot(energy_up, orbitals_up[:, orbital_index], label=f"{labels[index]} (Spin Up)",
                 color=colors[index], linewidth=line_thickness, linestyle='-')
        # Spin-down
        plt.plot(energy_down, orbitals_down[:, orbital_index], label=f"{labels[index]} (Spin Down)",
                 color=colors[index], linewidth=line_thickness, linestyle='--')

    # Customize the plot
    plt.xlabel('$E-E_F$ (eV)', fontsize=font_size)
    plt.ylabel('Partial Density of States', fontsize=font_size)
    plt.title('Spin-Polarized Partial Density of States', fontsize=font_size)

    # Set axis limits if provided
    if x_limits:
        plt.xlim(x_limits)
    if y_limits:
        plt.ylim(y_limits)

    # Customize tick parameters
    plt.tick_params(axis='both', which='major', labelsize=font_size)

    # Customize frame thickness
    plt.gca().spines['top'].set_linewidth(frame_thickness)
    plt.gca().spines['bottom'].set_linewidth(frame_thickness)
    plt.gca().spines['left'].set_linewidth(frame_thickness)
    plt.gca().spines['right'].set_linewidth(frame_thickness)

    # Add legend outside the plot frame
    plt.legend(fontsize=font_size, loc='upper left', bbox_to_anchor=(1, 1))

    # Adjust layout to make room for the legend
    plt.tight_layout(rect=[0, 0, 1, 1])

    # Show the plot
    plt.show()

# Example usage
filename_up = 'PDOS_Fe_UP.dat'  # Replace with your spin-up .dat file
filename_down = 'PDOS_Fe_DW.dat'  # Replace with your spin-down .dat file

# Read data from files
energy_up, orbitals_up = read_partial_dos_data(filename_up)
energy_down, orbitals_down = read_partial_dos_data(filename_down)

# Customize the selection of orbitals to plot
# List of indices: 0 for 's', 1 for 'p_x', 2 for 'p_y', ..., 8 for 'd_x2_y2'
selected_orbitals = [0, 4, 5, 6, 7, 8]  # Example: Plot s and p orbitals (p_x, p_y, p_z)
labels = ['s', 'd_xy', 'd_yz', 'd_z2', 'd_xz', 'd_x2_y2']  # Labels for the selected orbitals
colors = ['blue', 'red', 'green', 'orange', 'brown', 'yellow']  # Colors for the selected orbitals

# Customize the plot parameters as needed
plot_spin_polarized_partial_dos(
    energy_up, orbitals_up, energy_down, orbitals_down,
    selected_orbitals, labels, colors,
    font_size=14, line_thickness=2.5, frame_thickness=2,
    figure_size=(10, 7), x_limits=(-8, 8), y_limits=(-2, 2)
)
```

## **Partial density of states - multiple elements**

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to read partial DOS data from a .dat file
def read_partial_dos_data(filename):
    # Assuming the .dat file has the following columns:
    # energy, s, p_x, p_y, p_z, d_xy, d_yz, d_z2, d_xz, d_x2_y2
    data = np.loadtxt(filename)
    energy = data[:, 0]
    orbitals = data[:, 1:]  # Extract all orbital columns
    return energy, orbitals

# Function to plot partial DOS for multiple elements
def plot_multiple_partial_dos(
    filenames, selected_orbitals, labels_per_element, colors_per_element,
    font_size=12, line_thickness=2, frame_thickness=1.5, figure_size=(10, 7),
    x_limits=None, y_limits=None
):
    # Set figure size
    plt.figure(figsize=figure_size)

    # Loop over each file and plot the selected orbitals
    for i, filename in enumerate(filenames):
        # Read the data from the file
        energy, orbitals = read_partial_dos_data(filename)

        # Plot the selected orbitals for this element
        for j, orbital_index in enumerate(selected_orbitals[i]):
            plt.plot(
                energy, orbitals[:, orbital_index],
                label=f"{labels_per_element[i][j]} ({filename})",
                color=colors_per_element[i][j], linewidth=line_thickness
            )

    # Customize the plot
    plt.xlabel('Energy (eV)', fontsize=font_size)
    plt.ylabel('Partial Density of States', fontsize=font_size)
    plt.title('Partial Density of States for Multiple Elements', fontsize=font_size)

    # Set axis limits if provided
    if x_limits:
        plt.xlim(x_limits)
    if y_limits:
        plt.ylim(y_limits)

    # Customize tick parameters
    plt.tick_params(axis='both', which='major', labelsize=font_size)

    # Customize frame thickness
    plt.gca().spines['top'].set_linewidth(frame_thickness)
    plt.gca().spines['bottom'].set_linewidth(frame_thickness)
    plt.gca().spines['left'].set_linewidth(frame_thickness)
    plt.gca().spines['right'].set_linewidth(frame_thickness)

    # Add legend
    plt.legend(fontsize=font_size)

    # Show the plot
    plt.show()

# Example usage
filenames = ['element1_pdos.dat', 'element2_pdos.dat']  # Replace with your .dat file names

# Customize the selection of orbitals for each element
# Example: For element1, plot s and p_x; for element2, plot p_y and d_z2
selected_orbitals = [
    [0, 1],  # Indices for orbitals in element1 (s, p_x)
    [2, 6]   # Indices for orbitals in element2 (p_y, d_z2)
]

# Labels for each orbital of each element
labels_per_element = [
    ['s (Element 1)', 'p_x (Element 1)'],
    ['p_y (Element 2)', 'd_z2 (Element 2)']
]

# Colors for each orbital of each element
colors_per_element = [
    ['blue', 'red'],
    ['green', 'orange']
]

# Customize the plot parameters as needed
plot_multiple_partial_dos(
    filenames, selected_orbitals, labels_per_element, colors_per_element,
    font_size=14, line_thickness=2.5, frame_thickness=2,
    figure_size=(12, 8), x_limits=(-10, 10), y_limits=(0, 50)
)
```
